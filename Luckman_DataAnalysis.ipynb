## Creating a new column and sorting data into it

So ideally I can create a new column in this data frame, then run a for loop that will look for certain naming conventions and sort the payees into one of 4 categories: "verizon", "spectrum", "optimum", "other".

I need to:
*   create a new column
*   create a for loop using elif functions

It looks like I may be able to combine these steps defining a new column via an operation on existing columns.

Actually the big challenge here is how do I run operations on a string to find my search terms...?

```python
#first I will make a new column
provider = []

#next I will make a for loop that searches the strings in "payee_name" for
#substrings that identify the major providers

for name in everything["payee_name"]:
    if "time" in name.lower():
        provider.append("Spectrum")
    elif "spectrum" in name.lower():
        provider.append("Spectrum")
    elif "verizon" in name.lower():
        provider.append("Verizon")
    elif "altice" in name.lower():
        provider.append("Optimum")
    elif "optimum" in name.lower():
        provider.append("Spectrum")
    else:
        provider.append("other")

everything["provider"] = provider
everything
```
